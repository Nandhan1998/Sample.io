```python
import seaborn as sea
import tensorflow as tf
```

```python
data=sea.load_dataset("iris")
data.head()
```

```python
data["species"].unique()
```

```python
# clean the data
dictionary1={"setosa":0,"versicolor":1,"virginica":2}
```

```python
data["species"]=data["species"].map(dictionary1)
```

```python
x=data.drop("species",axis=1)
y=tf.keras.utils.to_categorical(data["species"])
```

```python
# input_layer=tf.keras.layers.Dense(units=4,input_dim=(4,))
# layer2=tf.keras.layers.Dense(units=8)
# layer3=tf.keras.layers.Dense(units=16)
# layer4=tf.keras.layers.Dense(units=32)
# layer5=tf.keras.layers.Dense(units=64)
# layer6=tf.keras.layers.Dense(units=128)
# layer7=tf.keras.layers.Dense(units=128)
# layer8=tf.keras.layers.Dense(units=64)
# layer9=tf.keras.layers.Dense(units=32)
# layer10=tf.keras.layers.Dense(units=16)
# layer11=tf.keras.layers.Dense(units=8)
# out_put=tf.keras.layers.Dense(units=3)
input_layer = tf.keras.layers.Dense(units=4, input_dim=4, activation='relu')
layer2 = tf.keras.layers.Dense(units=8, activation='relu')
layer3 = tf.keras.layers.Dense(units=16, activation='relu')
layer4 = tf.keras.layers.Dense(units=32, activation='relu')
layer5 = tf.keras.layers.Dense(units=64, activation='relu')
layer6 = tf.keras.layers.Dense(units=128, activation='relu')
layer7 = tf.keras.layers.Dense(units=128, activation='relu')
layer8 = tf.keras.layers.Dense(units=64, activation='relu')
layer9 = tf.keras.layers.Dense(units=32, activation='relu')
layer10 = tf.keras.layers.Dense(units=16, activation='relu')
layer11 = tf.keras.layers.Dense(units=8, activation='relu')
out_put = tf.keras.layers.Dense(units=3, activation='softmax')
```

```python
model=tf.keras.Sequential()
```

```python
#  add layers to the model
model.add(input_layer)
model.add(layer2)
model.add(layer3)
model.add(layer4)
model.add(layer5)
model.add(layer6)
model.add(layer7)
model.add(layer8)
model.add(layer9)
model.add(layer10)
model.add(layer11)
model.add(out_put)
```

```python
# compile the model
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(x,y,epochs=30)
```

```python
# to check the prediction
import numpy as np
model.predict(np.array([[5.1,3.5,1.4,0.2]])).reshape([-1,1])
```
